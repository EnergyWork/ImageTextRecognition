```python
import numpy as np
import cv2
import os
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np
from skimage.util import random_noise
```

```python
cv2.IMREAD_GRAYSCALE
```

```python
img = cv2.imread('.//images_dataset1//Train//img1018.jpg', cv2.IMREAD_GRAYSCALE)
print('Image Dimensions :', img.shape, type(img))
#img = cv2.medianBlur(img,5)

img = random_noise(img, mode='gaussian', seed=42)
img = np.array(255*img, dtype = 'uint8')

#img = cv2.bilateralFilter(img, 7, 50, 50)

ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

#titles = ['Original Image', 'Global Thresholding (v = 127)', 'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]
```

```python
blur = cv2.GaussianBlur(img, (5,5), 0)
ret4, th4 = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
path = './/output//outGaussBlur.jpeg'
save = cv2.imwrite(path, th4)
plt.imshow(th4, 'gray')
```
